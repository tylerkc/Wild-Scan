```python
# load master metadata from repo
# (same initial step for new dataset)
grp_meta_df =[]
grp_meta_df = pd.read_csv(f's3://aai-540-data/annotations/grp4_metadata_with_datetime.csv')
grp_meta_df['date_captured'] = pd.to_datetime(grp_meta_df['date_captured'], errors='coerce')
grp_meta_df.head()
```

```python
len(grp_meta_df)
```

```python
# Add Time Feature Columns ( Year-Month, Minute-Of-Day, Day-of-Year, Cyclical Encodings of MinuteOfDay and DayOfYear)

grp_meta_df['year_month'] = grp_meta_df['date_captured'].dt.to_period('M')

# Extract minute of day from 'date_captured' column
grp_meta_df['minute_of_day'] = grp_meta_df['date_captured'].dt.hour * 60 + grp_meta_df['date_captured'].dt.minute

# Extract day of year
grp_meta_df['day_of_year'] = grp_meta_df['date_captured'].dt.dayofyear

# Apply cyclical encoding of minute of day to expose nocturnal habits
grp_meta_df['minOfDay_sin'] = np.sin(2 * np.pi * grp_meta_df['minute_of_day'] / 1440)	
grp_meta_df['minOfDay_cos'] = np.cos(2 * np.pi * grp_meta_df['minute_of_day'] / 1440)

# Apply cyclical encoding of day of year to expose seasonal changes in animal habits
grp_meta_df['dayOfYear_sin'] = np.sin(2 * np.pi * grp_meta_df['day_of_year'] / 365)
grp_meta_df['dayOfYear_cos'] = np.cos(2 * np.pi * grp_meta_df['day_of_year'] / 365)
```

```python
# SPLIT I: Locations-based
# split df according to 'data_split' column. 
# 'train' samples are to be used for training and validation
# 'val' samples are to be used as production.
# spatial and temporal dataset splits are both going to be included later.
df = grp_meta_df.copy() # replace code later using csv file generated by athena query
df = df.reset_index(drop=True)

# Create splits
known_locs_df = df[df['split_type'] == 'train']
new_locs_df = df[df['split_type'] == 'val']
```

```python
# show known and new locs df value counts in one table
known_new_counts = pd.concat([known_locs_df['label'].value_counts(), new_locs_df['label'].value_counts()], axis=1).fillna(0).astype(int)
known_new_counts.columns = ['Known Locs', 'New Locs']
known_new_counts.loc['Total'] = known_new_counts.sum(numeric_only=True)
display(known_new_counts)
known_new_counts = known_new_counts.drop('Total')
```

```python
import seaborn as sns
import matplotlib.pyplot as plt


plt.figure(figsize=(12, 6))

sns.set_palette("tab10")
ax = sns.barplot(data=known_new_counts.reset_index().melt(id_vars='label'), 
            x='label', y='value', hue='variable')
# Set the title and labels
plt.title('Annotated Animal Class Distribution Based on Location Type')
plt.xlabel('Location')
plt.ylabel('Count')
plt.xticks(rotation=45)

# configure legend to add the total count per location

legend_labels = [f'Known : {known_new_counts['Known Locs'].sum()}',
                  f'New : {known_new_counts['New Locs'].sum()}'] 

handles, labels = ax.get_legend_handles_labels()

ax.legend(title='Location Type', loc='upper right', handles = handles, labels=legend_labels)
#plt.legend(title='Dataset Type', loc='upper right')

plt.tight_layout()
plt.show()
```

```python
# now display the % distribution of the classes in known and new locations
tmp_df = known_new_counts.T.copy()
known_new_counts_percent = tmp_df.div(tmp_df.sum(axis=1), axis=0) * 100


plt.figure(figsize=(6, 6))
sns.set_palette("tab20", n_colors=len(known_new_counts))

ax = known_new_counts_percent.plot(kind='bar', stacked=True, figsize=(14, 14))

# Melt the DataFrame for easier plotting
#ax = sns.barplot(data=known_new_counts_percent.reset_index().melt(id_vars='index'), 
#            x='index', y='value', hue='variable')
# Set the title and labels
plt.title('Annotated Animal Class Distribution Based on Location Type (Percentage)')
plt.xlabel('Location')
plt.ylabel('Percentage (%)')
plt.xticks(rotation=0)

plt.legend(title='Dataset Type', loc='upper right')

plt.tight_layout()
plt.show()

```

```python
# split 'known locations df' into train-val-test sets common to THREE IMAGE CLASSIFIERS for initial model training, evaluation, baseline performance, monitoring

# new 'train' set: filter known_locs_df by
train_year_month_start = known_locs_df['year_month'].min()
train_year_month_end = train_year_month_start + 12

# set temporal filters for the validation set
val_year_month_start = train_year_month_end + 1
val_year_month_end = val_year_month_start + 11

# create the train dataset from the known locations df
# TRAIN SET: Establishes KNOWN LOCATIONS, KNOWN LABELS, KNOWN TEMPORAL FEATURE RANGE
train_df = known_locs_df[known_locs_df['year_month'] <= train_year_month_end]

# extract all labels known to this training set
train_labels = train_df['label'].unique()
label_map = {label: idx for idx, label in enumerate((train_labels))}
print("labels for training")
display(label_map)

# create the validation dataset from the known locations df
# VAL SET: should be of SAME LOCATIONS, SAME LABELS, and SAME TEMPORAL FEATURE RANGE as the TRAIN SET
val_df = known_locs_df[(known_locs_df['year_month'] >= val_year_month_start) & (known_locs_df['year_month'] <= val_year_month_end)]
val_df = val_df[val_df['label'].isin(train_labels)]

print(f"\ntrain set:")
print(f"\ttime range: '{train_year_month_start}' to'{train_year_month_end}'")
print(f"\tnum samples: {len(train_df)}")

print(f"validation set:")
print(f"\ttime range: '{val_year_month_start}' to'{val_year_month_end}'")
print(f"\tnum samples: {len(val_df)}")
```

```python
# show train and val df value counts in one table
train_val_counts = pd.concat([train_df['label'].value_counts(), val_df['label'].value_counts()], axis=1).fillna(0).astype(int)
train_val_counts.columns = ['Train', 'Val']
train_val_counts.loc['Total'] = train_val_counts.sum(numeric_only=True)
train_val_counts
```

```python
# Configure Test Set
# set temporal filters for the validation set
test_year_month_start = val_year_month_end + 1

test_df = known_locs_df[known_locs_df['year_month'] >= test_year_month_start]
len(test_df)
```
